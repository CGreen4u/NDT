```python

#crop and image
import os

for filename in os.listdir(directory):
    if filename.endswith(".jpg"): 
        print(os.path.join(directory, filename))

        image = cv2.imread(filename)
        image = cv2.resize(image, (960, 540))
        gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        edged = cv2.Canny(image,100,200)
        threshold = cv2.threshold(gray, 155, 255, cv2.THRESH_BINARY)
        cnts, _ = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 
        idx = 0 
        os.chdir(r'X:\Miller_RL\Repairs\crop_files')
        for c in cnts: 
            x,y,w,h = cv2.boundingRect(c) 
            if w>50 and h>50: 
                idx+=1 
                new_img=image[y:y+h,x:x+w] 
                cv2.imwrite(str(idx) + filename + '.png', new_img) 
        cv2.imshow("im",image) 
        cv2.waitKey(0) 
        continue
    else:
        continue
# retval, threshold = cv2.threshold(grayscaled, 155, 255, cv2.THRESH_BINARY)
# cv2.imshow('original',img)
# cv2.imshow('threshold',threshold)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
